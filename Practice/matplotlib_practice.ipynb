```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot and add a label for the legend using MATLAB
plt.plot(x_axis, y_axis, label='Boston')

# Create the labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create the plot using MATLAB
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")

# Create the labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add the legend
plt.legend()
```

```python
# Create the plot with ax.plt() using Object-Oriented
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, label="Boston", color="green", marker="*", linewidth=2)

# Add a legend
ax.legend()

# Add a title and axes labels
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Fare ($)")
ax.set_xlabel("Date")

# Add a y limit
ax.set_ylim(0, 45)

# Add grid lines
ax.grid()
```

```python
# Create the plot with ax.plt() using Object-Oriented
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot using MATLAB
plt.barh(x_axis, y_axis, color="magenta", label="Boston")

#Add axes labels and title
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")

# Invert the y-axis so that January is at the top
plt.gca().invert_yaxis()

# Add a legend
plt.legend()
```

```python
# Create the plot with ax.plt() using Object-Oriented
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() using Object-Oriented
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")

# Add a legend
ax.legend()

#Add a title and axes labels
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Date")
ax.set_xlabel("Fare ($)")

#Switch the axis so that the Fare data is on the x-axis
ax.invert_yaxis()
```

```python
# Create the scatter plot using MATLAB
plt.plot(y_axis, x_axis, 'o', c="red", label="Chicago")

# Add a legend
plt.legend()

# Add a title and axes labels
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")

# Invert the y-axis so that January is at the top
plt.gca().invert_yaxis()
```

```python
# Create the bubble chart using MATLAB
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# adjust size of the markers
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
# create the bubble chart with larger markers
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# create the bubble chart with larger markers using List Comprehension
plt.scatter(x_axis, y_axis, s = [I * 3 for I in y_axis])
```

```python
# create a scatter plot using object-oriented
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a bubble chart using Object-Oriented
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s= [i * 5 for i in y_axis], c="skyblue", alpha=0.2, linewidth=2, label="Boston")

# Add a legend
ax.legend()
```

```python
# Create a pie chart using MATLAB
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
